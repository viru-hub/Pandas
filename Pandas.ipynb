# Merge

```python
data = pd.read_csv('cadbury.csv')
```

```python
data1 = pd.read_csv('alt_gg.csv')
data1
```

```python
merger_df = (data.merge(data1,on='id')) # meging data1 with data table with a common column= 'id'
```

```python
merger_df
```

```python
merger_df['taste_x'].describe()
```

```python
filterr = merger_df.loc[merger_df['quatity']==36,:]  #filter using .loc 
```

```python
filterr
```

```python
fltr = merger_df.loc[(merger_df['quatity']>30) & (merger_df['taste_y']>17)]
```

```python
fltr
```

```python
def tasteful(k):
    if k > 16:
        return 1
    else:
        return 0
    
merger_df['taste'] = merger_df.apply(lambda row:tasteful(row['taste_x']),axis=1)
```
